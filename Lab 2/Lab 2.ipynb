```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

%matplotlib inline
```

```python
img = mpimg.imread("2.jpeg")
print(img)
```

```python
plt.imshow(img)
```

```python
rgb_weights = [0.2989, 0.5870, 0.1140]

grayscale_image = np.dot(img[...,:3], rgb_weights)
print(grayscale_image)
```

```python
plt.imshow(grayscale_image, cmap=plt.get_cmap("gray"))
```

```python
plt.subplot(2,2,1)
plt.imshow(img)

plt.subplot(2,2,2)
plt.imshow(grayscale_image, cmap=plt.get_cmap("gray"))
```

## Take two images and perform Image subtraction, addition, and multiplication of those images using arrays

```python
first = mpimg.imread("1.jpeg")
second = mpimg.imread("2.jpeg")
```

```python
print(first)
```

```python
print(second)
```

```python
# SUBTRACTION

subt = np.subtract(first,second)
plt.imshow(subt)
```

```python
# ADDITION

add = np.add(first,second)
plt.imshow(add)
```

```python
# MULTIPLICATION

mult = np.multiply(first,second)
plt.imshow(mult)
```

## Converting from one color space into another (RGB to HSV conversion)

```python
from skimage.color import rgb2hsv
```

```python
rgb_img = mpimg.imread("1.jpeg")
plt.imshow(rgb_img)
```

```python
hsv_image = rgb2hsv(rgb_img)
```

```python
plt.subplot(2,2,1)
plt.title("Normal")
plt.imshow(rgb_img)

plt.subplot(2,2,2)
plt.title("HSV")
plt.imshow(hsv_image)
```

## Image resizing using bilinear, bicubic and nearest neighbour interpolation methods

```python
img = mpimg.imread("2.jpeg")
img = img[300:1100,:500,:]
plt.imshow(img)
```

```python
# LINEAR INTERPOLATION
import cv2
linear_img = cv2.resize(img,None,fx=10,fy=10,interpolation=cv2.INTER_LINEAR)
```

```python
plt.subplot(2,2,1)
plt.title("Normal")
plt.imshow(img)

plt.subplot(2,2,2)
plt.title("Linear Interploation")
plt.imshow(linear_img)
```

```python
cubic_img = cv2.resize(img,None,fx=10,fy=10,interpolation=cv2.INTER_CUBIC)
```

```python
plt.subplot(2,2,1)
plt.title("Normal")
plt.imshow(img)

plt.subplot(2,2,2)
plt.title("Bi-Cubic Interploation")
plt.imshow(cubic_img)
```

```python
 nearest_neighbour_img = cv2.resize(img,None,fx=10,fy=10,interpolation=cv2.INTER_NEAREST)
```

```python
plt.subplot(2,2,1)
plt.title("Normal")
plt.imshow(img)

plt.subplot(2,2,2)
plt.title("Nearest neighbour Interploation")
plt.imshow(nearest_neighbour_img)
```
