Creating a circular mask on the input image. The example of output image is below.
Hint: Slicing and masking with numpy arrays can be used to create a circular mask on
the input image. 

```python
img = cv2.imread('2.jpeg')
mask = np.zeros(img.shape, dtype = np.uint8)
mask = cv2.circle(mask, (260, 300), 225, (255, 255, 255), -1)

result = cv2.bitwise_and(img, mask)
result[mask == 0] == 255
```

```python
plt.subplot(2, 2, 1)
plt.imshow(img)

plt.subplot(2, 2, 2)
plt.imshow(result)
```

8. Drawing on an image in Python: <br />
a) You can draw lines or other geometric shapes on an image. For example, drawing
ellipse on input image and output will look like below.

```python
# Deawing an ellipse
from PIL import Image, ImageDraw
im = Image.open("2.jpeg")
x, y =  im.size
eX, eY = 100, 200 
ellipse =  (x/2 - eX/2, y/2 - eY/2, x/2 + eX/2, y/2 + eY/2)
draw = ImageDraw.Draw(im)
draw.ellipse(ellipse,fill=222)
del draw
im.save("output.png")
im
```

b) Drawing text on an image. Write some text on input image. Use a function in Python to change its font as well

```python
!pip install Python-fontconfig
```

```python
from PIL import Image,ImageDraw,ImageFont
import fontconfig

fonts = fontconfig.query(lang='en')
for i in range(1, len(fonts)):
    if fonts[i].fontformat == 'TrueType':
        absolute_path = fonts[i].file
        break


unicode_text = "This is a picture of a banana"
font = ImageFont.truetype(absolute_path, 28, encoding="unic")


text_width, text_height = font.getsize(unicode_text)

img = Image.open("2.jpeg")


# draw the text onto the text canvas, and use black as the text color
draw = ImageDraw.Draw(img)
draw.text((5,5), u'This is a picture of a banana', 'yellow', font)

# save the blank canvas to a file
img.save("unicode-text.png", "PNG")
display(img)
```

9. R,G,B channels splitting and merging. Read any RGB image as input, split three
channels R, G, B and display these channels/matrices as output images. Then, merge
the three channels again and display same RGB image.

```python
import cv2
import glob
import os
```

```python
img=cv2.imread('2.jpeg',1)
```

```python
b,g,r=cv2.split(img)
cv2_imshow(b)

cv2.destroyAllWindows()
```

```python
cv2_imshow(g)
cv2.destroyAllWindows()
```

```python
cv2_imshow(r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_combine=cv2.merge((b,g,r))
cv2_imshow(img_combine)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
