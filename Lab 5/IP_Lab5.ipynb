```python
import matplotlib.pyplot as plt
from skimage.transform import warp
from skimage import data
from skimage.transform import SimilarityTransform
from skimage import transform as tf
%matplotlib inline
```

```python
import numpy as np
img = plt.imread("bird.png")
img = np.asarray(img,dtype=np.float32)
img
```

```python
plt.imshow(img)
```

```python
import math
#text = data.text()

tform = tf.SimilarityTransform(translation=(img.shape[0]/2, -100))

bottom_left = tf.warp(img, tform)
top_right = tf.warp(bottom_left, tform.inverse)
```

```python
plt.imshow(bottom_left, cmap=plt.cm.gray)
```

```python
plt.imshow(top_right, cmap=plt.cm.gray)
```

b)	Scale, rotation, translation 

You can use the above functions for performing these three geometric transforms together.

About scaling:
(x, y) = (a*x, b*y) where (x, y) are new coordinates after scaling, (x, y) are original coordinates, a and b are scaling factors.


```python
tform_b = tf.SimilarityTransform(scale=2,rotation=math.pi/4, translation=(img.shape[0]/2, -100))
s_r_t = tf.warp(img, tform_b)
plt.imshow(s_r_t)
```

## 2.	Perform the following geometric transforms without using inbuilt functions.

```python
# a) Apply the horizontal shear (shear parallel to the x axis) on the image with shear factor = 2. 
# It converts the coordinates from (x,y) to the point (x+my, y) where m is the shear factor.

import cv2
img = cv2.imread('bird.png') 

rows, cols, ch = img.shape 

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1, pts2) 
dst = cv2.warpAffine(img, M, (cols, rows)) 

plt.imshow(dst)
```
